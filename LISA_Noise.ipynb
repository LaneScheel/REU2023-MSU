```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f  = np.linspace(0.0001,1,100000)
fa = 0.0004
fb = 0.025

A = 1.28e-44
alpha = 1.63
a1 = -0.224
b1 = -2.704
ak = -0.361
bk = -2.378

T = 4

f1 = 10**((a1*np.log10(T))+b1)
f2 = 10**(-3.318)
fk = 10**((ak*np.log10(T))+bk)
```

```python
S1 = 5.76e-48 * (1 + ((fa/f)**2))     #s^-4 Hz^-1
S2 = 3.60e-41                         #Hz^-1
R  = 1 + ((f/fb)**2)                  #unitless
Sc = (A/2)*(np.e**(-(f/f1)**alpha))*(f**(-7/3))*(1+np.tanh((fk-f)/f2))
```

```python
S_noise = f*(((1/2) * (20/3) * ((S1/((2*np.pi*f)**4)) + S2) * R) + Sc)
```

```python
h_c = np.sqrt(S_noise)
```

```python
plt.plot(f, h_c, color='black', linestyle='dashed', label='LISA Noise + WD Conf.')
plt.xlabel('Frequency [Hz]')
plt.ylabel('h(c) (strain)')
plt.xlim(1e-4, 1e-1)
plt.ylim(1e-22, 1e-17)
plt.xscale('log')
plt.yscale('log')
plt.suptitle('Barack and Cutler Reproduction')
plt.title('Assuming Only Circular Binary Orbits')
plt.legend()
plt.grid()
plt.show()
```
