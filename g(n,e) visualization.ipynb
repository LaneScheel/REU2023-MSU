# Summarization of Results
### Harmonic n = 2 appears to clearly dominate at e = 0
##### Harmonic n = 1 stays very flat relative to the others as we sweep through the eccentricities
##### Harmonic n = 2 decays starts high as we sweep through the eccentricities
##### Harmonic n = 3 grows quickly relative to the others as we sweep through the eccentricities
##### Harmonic n = 4 grows grows at a rate between that of n = 1 and n = 3 as we sweep through the eccentricities

```python
import numpy as np
from scipy.special import jv
import matplotlib.pyplot as plt
```

```python
#Define the eccentiricity space
e = np.linspace(0, 0.9, 2000)
```

```python
#Define relative power radiated function
def g(n):
    g = ((n**4)/32)*((((jv(n-2,(n*e)))-(2*e*(jv(n-1,(n*e))))+((2/n)*(jv(n,(n*e))))+(2*e*(jv(n+1,(n*e))))-((jv(n+2,(n*e)))))**2)+(((1-(e**2))*((jv(n-2,(n*e)))-(2*(jv(n,(n*e)))+(jv(n+2,(n*e))))))**2)+((4/(3*(n**2)))*(((jv(n,(n*e))))**2)))
    return g
```

```python
#Plot relative power radiated into each harmonic vs eccentriicity
plt.plot(e, g(1), 'r-', label='n=1')
plt.plot(e, g(2), 'y-', label='n=2')
plt.plot(e, g(3), 'g-', label='n=3')
plt.plot(e, g(4), 'c-', label='n=4')
plt.plot(e, g(5), 'b-', label='n=5')
plt.plot(e, g(6), 'm-', label='n=6')
plt.title('Relative Power Rediated into a Harmonic by Eccentricity')
plt.xlabel('Eccentricity')
plt.ylabel('g(n,e)  relative power radiated')
plt.legend()
plt.show()
```
